# Make Theory Plots
Make some nice math-y plots to include in the methods section.

```python
import numpy as np
import scipy.special
import matplotlib.pyplot as plt
import seaborn
seaborn.set_context(context='paper',font_scale=2.5)
%matplotlib inline
```

Compute the Bessel functions of the first kind according to the equation,
$$
J_{\alpha}(x) = \sum_{m=0}^{\infty}\frac{(-1)^m}{m!\Gamma(m+\alpha+1)}\left(\frac{x}{2}\right)^{2m+\alpha}.
$$
To do this, we'll use the [`scipy.special` module](http://docs.scipy.org/doc/scipy/reference/special.html).

```python
orders = [0,1,2,3,4]
bessels = {}
x = np.linspace(0,20,100)
for o in orders:
    bessels['J'+str(o)] = scipy.special.jv(o,x)
```

Finally, build and save the plot.

```python
fig = plt.figure(figsize=(8,8))
ax = fig.gca()
for key,o in zip(bessels,orders):
    ax.plot(x,bessels[key],label=r'$J_{%d}$'%(o),linewidth=3)
ax.set_ylabel(r'$J_{\alpha}(x)$')
ax.set_xlabel(r'$x$')
ax.set_title(r'Bessel functions of the first kind')
ax.legend(loc='best')
plt.savefig(__dest__,format='pdf',dpi=1000)
plt.show()
```
