# Make Sample Simulation Data
Here is some data generated by some python program. We'll export this as a file and then it will be read in to another notebook.

```python
import numpy as np
import pickle
```

First, generate a bunch of sample data from a normal distribution.

```python
N=10000
p = np.random.normal(loc=0.0,scale=1.5,size=N)
```

Now, make some random noise

```python
noise_level = 0.1*np.max(p)
p_noise = np.random.rand(N)*(2.*noise_level) - noise_level
```

And add it into the distribution

```python
p_noisy = p + p_noise
```

Finally, print all of this to a pickle file.

```python
with open(__dest__,'wb') as f:
    pickle.dump(p_noisy,f)
```
